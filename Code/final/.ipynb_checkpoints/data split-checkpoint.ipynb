## Import Statments

```python
import copy
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import seaborn as sns

r = 1337
random.seed(r)
np.random.seed(r)
```

## Load Data

```python
data = pd.read_json('../../Data/data_impute.json')
labels = data[['label']]
data.drop(columns='label', inplace=True)
```

## Split Data

```python
train_test_size = 0.8

X_train_test_idx, X_hold_idx, y_train_test, y_hold = train_test_split(
    np.arange(len(data)),
    labels.values.ravel(),
    train_size=train_test_size,
    random_state=r,
    shuffle=True,
    stratify=labels.values.ravel()
)
```

```python
X_train_test = data.iloc[X_train_test_idx]
X_hold = data.iloc[X_hold_idx]
```

## Compute Correlation

### Impute Data

```python
# remove correlated features according to train-test split only
imp = IterativeImputer(max_iter=100, random_state=r)

X_train_test_imp = pd.DataFrame(
    imp.fit_transform(X_train_test.values),
    columns=X_train_test.columns,
    index=X_train_test.index
)
```

### Correlate

```python
corr1 = X_train_test_imp.corr()
plt.subplots(figsize=(20, 20))
sns.set(font_scale=1.5)

ax = sns.heatmap(
    corr1,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

### Choose Correlated Features to Remove

```python
corr_thresh = 0.90  # threshold for correlation. for any two variables with correlation > thresh, one is removed

thresh = corr1.abs() > corr_thresh

keep = copy.deepcopy(data.columns).to_list()

print('Removed features: ')
# keep features whose correlation with other features is <= corr_thresh
for i in range(0, len(thresh.index)):
    for j in range(i+1, len(thresh.columns)):
        if thresh.iloc[i, j]:
            if thresh.columns[j] in keep:
                print('\t', thresh.columns[j])
                keep.remove(thresh.columns[j])
```

```python
corr2 = X_train_test_imp[keep].corr()
plt.subplots(figsize=(20, 20))
sns.set(font_scale=1.5)

ax = sns.heatmap(
    corr2,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
ax.set_title('Remaining Features After Correlated Variables were Removed');
```

## Save Data

```python
X_train_test.insert(loc=0, column='label', value=y_train_test)
X_train_test[keep].to_json('./data/train-test.json')    # save train-test split with correlated features removed

X_hold.insert(loc=0, column='label', value=y_hold)
X_hold[keep].to_json('./data/holdout.json')                   # save holdout split with correlated features removed
```
