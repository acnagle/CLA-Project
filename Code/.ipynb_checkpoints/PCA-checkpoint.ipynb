```python
from sklearn import decomposition
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
np.set_printoptions(threshold=np.inf)  # prints a full matrix rather than an abbreviated matrix

# define data and destination paths
dest_path = "/Users/Alliot/Documents/CLA-Project/Data/all-data-no-na/neural-network/"
data_path = "/Users/Alliot/Documents/CLA-Project/Data/data-sets/"
data_set = "data_2017_summer"

# if dest_path does not exist, create it
if not os.path.exists(dest_path):
    try:
        os.makedirs(dest_path)
    except OSError as e:
        if e.errno != errno.EEXIST:
            raise

# load data sets
X = np.load(data_path + data_set + ".npy")
y = np.load(data_path + data_set + "_labels.npy")
```

```python
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

X = (X - mean) / std

pca = decomposition.PCA(n_components=2, svd_solver='auto')
pca.fit(X)
```

```python
comp = pca.components_
s_vals = pca.singular_values_
pca_X = np.zeros(shape=(X.shape[0], pca.n_components))

for i in range(X.shape[0]):
    for j in range(comp.shape[0]):
        pca_X[i, j] = np.dot(comp[j], X[i, :]) * np.sqrt(s_vals[j])
```

```python
no_alg_idx = []
alg_idx = []

for i in range(X.shape[0]):
    if y[i] == 0:
        no_alg_idx.append(i)
    if y[i] > 0:
        alg_idx.append(i)
        
no_alg = pca_X[no_alg_idx, :]
alg = pca_X[alg_idx, :]

plt.figure()
plt.scatter(no_alg[:, 0], no_alg[:, 1], c='g')
plt.scatter(alg[:, 0], alg[:, 1], c='r')
plt.grid(True);
```
